```python
import string
from string import digits
#import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import gensim
from gensim.utils import simple_preprocess
import re
import nltk
nltk.download('stopwords')
stop_words = stopwords.words('english')
```

```python
def my_preprocessor(doc):
  doc_lower = doc.lower().replace("amp", "")
  doc_lower = re.sub(r'http\S+', '', doc_lower)
  remove_digits = str.maketrans('', '', string.digits)
  translator = str.maketrans('', '', string.punctuation)
  doc_no_punctuation = doc_lower.translate(translator).translate(remove_digits)
  return(doc_no_punctuation)
```

```python
def my_tokenizer2(doc):
  if len(doc) == 0:
    return('')
  else:
    doc_tokens =gensim.utils.simple_preprocess(doc, deacc=True)
    ps = PorterStemmer()
    doc_filtered = [w for w in doc_tokens if not w in stop_words]
    doc_stem = [ps.stem(w) for w in doc_filtered]
    # return(doc_stem)
    return (" ".join(doc_stem))
```

```python
import pandas as pd
import re
import string
from string import digits
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import gensim
from gensim.utils import simple_preprocess
import numpy as np


docs = pd.read_csv("/content/gdrive/MyDrive/aYin/landscape/ludwig/data/landscape_restoration_trees.tsv", sep="\t")
docs['message'] = docs['text'].apply(my_preprocessor)
docs['message'].replace('', np.nan, inplace=True)
docs.dropna(subset=['message'], inplace=True)
docs['message'] = docs['message'].apply(my_tokenizer2)
docs['message'].replace('', np.nan, inplace=True)
docs.dropna(subset=['message'], inplace=True)
print(len(docs))
docs['message2'] = docs['text'].apply(my_preprocessor)
docs['message2'].replace('', np.nan, inplace=True)
docs.dropna(subset=['message2'], inplace=True)
# docs.to_csv("train/landscape_restoration_trees_train_tokenized.tsv", sep="\t", index=False)
docs.to_csv("/content/gdrive/MyDrive/aYin/landscape/ludwig/data/landscape_restoration_trees_tokenized.tsv", sep="\t", index=False)
```
